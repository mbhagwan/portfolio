## Data Analysis and Visualization with Power BI Desktop

### Project Brief

AdventureWorks, a fictional global manufacturing company, produces cycling equipment and accessories.

The management needs a way to __track KPIs__ (sales, revenue, profit, returns), compare **regional**  

**performance**, **analyze product-level trends**, and **identify high-value customers.**

The folder AdventureWorks Raw Data, contains raw csv files with information about

transactions, returns, products, customers, and sales territories.

### Objective

Use Power BI Desktop to:

* Connect and transform the raw data

* Build a relational data model

* Create calculated columns and measures with DAX

* Design an interactive dashboard and visualize and analyze the data

### Connect and transform the raw data

First, connect and load the source data files found in the *AdventureWorks Raw Data* folder.

The imported data is then shaped and transformed in the Power Query Editor before being added to the   

data model for analysis.

In addition to the above, a date table is created in Power Query. This date table will be added to

the data model as it is a prerequisite to work with Data Analysis Expressions (DAX) time intelligence

functions.

### Build a relational data model

Once the data has been loaded into the data model it is necessary to create relationships between the  

tables to build a proper relational model before the data can be analyzed.

### Create calculated columns and measures with DAX

In the next step, DAX is used to create calculated columns, which are new columns of data, inside the  

existing tables. Calculated columns are used for filtering.

DAX is also used to create measures, which are then used in visuals to show aggregations like sum,   

count, or average.

### Design an interactive dashboard and visualize and analyze the data


In the final step, data visualization best practices are applied to design effective reports and  

dashboards. Various formatting and filtering options, bookmarks and page navigation tools, field 

and value parameters, and custom tooltips are used to enhance report interactivity and analysis.

The completed Power BI report consists of the following pages and screenshots are shown below:

* Executive Dashboard page to track KPIs (sales, revenue, profit, returns)

* Sales Territories page to compare regional performance

* Product Detail page to analyze product-level trends

* Customer Detail page to identify high-value customers
